# Recipe 2-9. Exploring Text Data

```python
import nltk
from nltk.corpus import webtext

wt_sentence = webtext.sents('firefox.txt')
wt_word = webtext.words('firefox.txt')
```

```python
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import string

```

```python
frequency_dist = FreqDist(wt_word)
```

```python
frequency_dist.plot()
```

```python
lardword = dict([(k,v) for k,v in frequency_dist.items() if len(k)>3])
frequency_dist = nltk.FreqDist(lardword)
frequency_dist.plot(50)
```

```python
## Step 9-6. Build a word cloud

!pip install wordcloud
```

```python
from wordcloud import WordCloud
wcloud = WordCloud().generate_from_frequencies(frequency_dist)
import matplotlib.pyplot as plt
plt.imshow(wcloud)
```

```python
## removing stop word

stop = stopwords.words('english')
wt = [x for x in wt_word if x not in stop]
```

```python
freq = nltk.probability.FreqDist(wt)
wordCl = WordCloud().generate_from_frequencies(freq)
plt.imshow(wordCl)
plt.axis(False);
```
